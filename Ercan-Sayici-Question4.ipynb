```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()

rdd=sc.textFile("/home/ercan/Desktop/hadoop/datasets/SayisalLoto.txt")

rdd1= rdd.flatMap(lambda line: line.split("}{"))
rdd2=rdd1.map(lambda line :line.split(" "))
rdd3=rdd2.map(lambda line :line[41])
rdd4=rdd3.map(lambda line :line.replace(",","").replace('"',""))
#rdd4.take(3)

def lit_to_tuple(arr):
    a1 = int(arr[0])
    a2 = int(arr[1])
    a3 = int(arr[2])
    a4 = int(arr[3])
    a5 = int(arr[4])
    a6 = int(arr[5])
    
    return [a1,a2,a3,a4,a5,a6]

rdd5 = rdd4.map(lambda x: x.split('#'))
rdd6 = rdd5
rdd7 = rdd6.cartesian(rdd6)
#rdd7.take(3)

def count(arr1,arr2):
    i = 0
    for a1 in arr1:
        for a2 in arr2:
            if a1 == a2:
                i += 1
    if i==3:
        return True
    else:
        return False
    
def count_arr(arr1,arr2):
    i = 0
    arr = []
    for a1 in arr1:
        for a2 in arr2:
            if a1 == a2:
                i += 1
                arr.append(a1)
    if i==3:
        return arr
    
rdd7 = rdd7.filter(lambda x:count(x[0],x[1])==True)

rdd7 = rdd7.map(lambda x: (count_arr(x[0],x[1]), x[0], x[1]))
#rdd7.take(3)

def sort_num(arr):
    a1 = int(arr[0])
    a2 = int(arr[1])
    a3 = int(arr[2])
    
    z = [a1, a2, a3]
    z.sort(reverse=True)
    fnl = []
    for a in z:
        fnl.append(str(a))
    return fnl

def sort_num_str(arr):
    a1 = int(arr[0])
    a2 = int(arr[1])
    a3 = int(arr[2])
    
    z = [a1, a2, a3]
    z.sort(reverse=True)
    fnl = str(z[0])
    fnl = fnl + '-' + str(z[1])
    fnl = fnl + '-' + str(z[2])
    
    return fnl

rdd7 = rdd7.map(lambda x: (sort_num_str(x[0]), x[1], x[2]))

rdd8 = rdd7.map(lambda x: x[0])

rdd9 = rdd8.map(lambda x : (x,1))

rdd9 = rdd9.reduceByKey(lambda x,y: x+y)
#rdd9.take(3)

rdd10 = rdd9.map(lambda x:(x[0], x[1]/2))

rdd11 = rdd10.sortBy(lambda x: x[1], ascending=False)

rdd11.take(3)
```
